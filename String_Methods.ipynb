# 1. String.capitalize()

```python
string = "python"
print(string.capitalize()) #First letter capital, rest small
```

# 2. String.casefold()

```python
string = "python"
print(string.casefold()) #is used to convert a string into lowercase in a more aggressive way than lower().
```

# 3. String.center()

```python
text = "hello"
print(text.center(10)) #is used to align a string in the center of a specified width by adding padding on both sides.
```

# 4. String.count()

```python
text = "hello"
print(text.count("l")) #count how many times a substring appears in a string.
text = "banana"
print(text.count("a", 2, 6)) #It counts only between index 2 and 6.
```

# 9. String.format()

```python
age = 21
print("My name is {} and I am {} years old".format("Shashank", age)) #insert values into a string at specified positions (string formatting).
```

# 10. String.format_map()

```python
data = {"name": "Shashank", "age": 21}
text = "My name is {name} and I am {age} years old"

print(text.format_map(data)) #is used to format a string using a dictionary (mapping).
```
