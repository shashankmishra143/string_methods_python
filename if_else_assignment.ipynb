```python
#SOLUTION-1

units = int(input("Enter number of units consumed :"))
total_bill = 0

if units <= 100:
    total_bill = 0

if units > 100 and units <= 200 :
    total_bill = (units - 100) * 5

if units > 200 :
    total_bill = 100 * 5 + (units - 200) * 10

print("Total electricity bill : ",total_bill)
```

```python
#SOLUTION-3

person_1_age = int(input("Enter the age of first person : "))
person_2_age = int(input("Enter the age of second person : "))
person_3_age = int(input("Enter the age of third person : "))
person_4_age = int(input("Enter the age of fourth person : "))

youngest = person_1_age
if person_2_age < youngest :
    youngest = person_2_age
if person_3_age < youngest :
    youngest = person_3_age
if person_4_age < youngest :
    youngest = person_4_age

print(f"Youngest one is {youngest} year old")
    
```

```python
#SOLUTION-4

salary = float(input("Enter the salary :"))
years_of_service = int(input("Enter the years of service : "))

net_bonus_amount = 0

if years_of_service > 10 :
    net_bonus_amount = (salary * 10) / 100
elif years_of_service >= 6 and years_of_service <=10 :
    net_bonus_amount = (salary * 8) / 100
else :
    net_bonus_amount = (salary * 5) / 100


    
print(f"The net bonus amount is {net_bonus_amount} rupees")
```

```python
#SOLUTION-6

marked_price = float(input("Enter the marked price : "))

if marked_price > 10000 :
    discount = (marked_price * 20) / 100
    net_amount = marked_price - discount
elif marked_price > 7000 and marked_price <= 10000 :
    discount = (marked_price * 15) / 100
    net_amount = marked_price - discount
else :
    discount = (marked_price * 10) / 100
    net_amount = marked_price - discount
print("Net amount : ",net_amount)    
    
```
